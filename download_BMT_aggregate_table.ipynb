```python
roadmap_2_0_adults = roadmap_2_0_BMT_roadmapIDs[roadmap_2_0_BMT_roadmapIDs.cohort != 'Peds']
roadmap_dyads = roadmap_2_0_adults[roadmap_2_0_adults['Unnamed: 5'].isna()]
roadmap_BMT_IDs_final =  pd.concat([roadmap_dyads.pt_rm_access_code, roadmap_dyads.cg_rm_access_code],ignore_index=True)
roadmap_BMT_IDs_final[335]
```

```python
# Enfore the DSN String has the service_name field instead of the SID
dsnStr = cx_Oracle.makedsn("abcd.umich.edu", "1521", service_name="ABCD.WORLD")
connection = oracledb.connect(
    user="username",
    password="password",
    dsn=dsnStr
)
cursor = connection.cursor()
print("Successfully connected to Oracle Database")
```

```python
df_BMT_aggregates_table = pd.DataFrame()
```

```python
for i in range(len(roadmap_BMT_IDs_final)):
# for i in range(5):    
       print("current participant being downloaded is", roadmap_BMT_IDs_final[i])
       cursor.execute(
        "select *from ROADMAP.STUDY_METRIC_AGGREGATES where STUDY_PRTCPT_ID = '" + str(roadmap_BMT_IDs_final[i]) + "'"
       )
       unique_strings = cursor.fetchall()
       column_names   = [i[0] for i in cursor.description]
       df_current = pd.DataFrame(unique_strings,columns=column_names)
       df_BMT_aggregates_table = pd.concat([df_BMT_aggregates_table,df_current],ignore_index= True)
    
```
